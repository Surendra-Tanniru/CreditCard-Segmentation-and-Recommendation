```python
# Select only the numeric columns from the dataset
nba_numeric = df[distance_columns]
# Normalize all of the numeric columns
nba_normalized = (nba_numeric - nba_numeric.mean()) / nba_numeric.std()
```

```python
import math
def euclidean_distance(row):
    """
    A simple euclidean distance function
    """
    inner_value = 0
    for k in distance_columns:
        inner_value += (row[k] - selected_card[k]) ** 2
    return math.sqrt(inner_value)

# Find the distance from each card in the dataset to selected card.
card_distance = df.apply(euclidean_distance, axis=1)
```

```python
# Finding the nearest neighbor
from scipy.spatial import distance

# Fill in NA values in nba_normalized
nba_normalized.fillna(0, inplace=True)

# Find the normalized vector for selected card.
# lebron_normalized = nba_normalized[df["BankName-CardVariant"] == "HDFC-6E Rewards-IndiGo HDFC Bank"]
card_normalized = nba_normalized[df["BankName-CardVariant"] == "SBI-Air India SBI Signature Card"]


# Find the distance between two cards and every other card.
euclidean_distances = nba_normalized.apply(lambda row: distance.euclidean(row, card_normalized), axis=1)
```

```python
# Create a new dataframe with distances.
distance_frame = pd.DataFrame(data={"dist": euclidean_distances, "idx": euclidean_distances.index})
```

```python
distance_frame
```

```python
distance_frame.sort_values("dist", inplace=True)
# Find the most similar card to selected card(the lowest distance to selected card is selected card, the second smallest is the most similar non-selected card)
second_smallest = distance_frame.iloc[1]["idx"]
most_similar_to_card = df.loc[int(second_smallest)]["BankName-CardVariant"]
```

```python
from sklearn.neighbors import KNeighborsRegressor
# Create the knn model.
# Look at the five closest neighbors.
knn = KNeighborsRegressor(n_neighbors=5,p=1)
```

```python
# Fit the model on the data.
knn.fit(df[x_columns], df[y_column])
```

```python
print ("The MSE is:",format(np.power(df[y_column]-knn.predict(df[x_columns]),2).mean()))
```

```python
# Make point predictions on the dataset using the fit model.
predictions = knn.predict(df[x_columns])
```

```python
predictions.shape
```

```python
predictions_df = pd.DataFrame(predictions)
predictions_df.columns=['idx_pred']
predictions_df

m=(predictions_df.dtypes=='float')
predictions_df.loc[:,m]=predictions_df.loc[:,m].astype(int)
predictions_df
```

```python
predictions_df.head(1)
bank_df.head(1)

d = {}
for a, b in zip(bank_df['idx'], bank_df['Bank-Card']):
    if isinstance(a, list):
        for c in a:
            d[c] = b
    else:
        d[a] = b

predictions_df['Bank-Card'] = predictions_df['idx_pred'].map(d)
predictions_df
```
