```python
## Aggregate information at the store level

df_pivot = pd.pivot_table(df, values=['AnnualFee', 'MinimumSpendsForAnnualFeeReversal', 'No_of_Dom_AirportLoungeBenefits',
                                      'No_Of_Int_AirportLoungeBenefits', 'FuelSurchargeWaiver_in_Percent',
                                      'FuelDiscount/Cashback_in_Percent','Maximum_FuelDiscount/CashbackinRupees',
                                      'FuelRewardPoints',      'MinimumSpendinRupees', 'MaximumDiscount/WaiverinRupees',
       'E-Commerce Cashback in percentage', 'Merchant Cashback in percentage', 'Mobile Recharge/wallet Cashback in percentage',
       'Domestic Air Ticket Cashback in percentage', 'Movie Ticket Cashback in percentage', 'Utility Cashback in percentage',
       'Domestic Travel in Percentage', 'Domestic Hotel spend Cashback in percentage',
       'Other Cashback in percentage', 'Cashback in points','Payback points for online spend',
                                      'Payback points for Utility spend', 'Payback points for dining/Hotel',
                                      'Payback points on Fuel Spend',
       'Payback points on Domestic Travel spend',       'Payback points for Annual spend limit',
       'domestic Air Ticket cashback in rupees',       'International Air Ticket cashback in rupees',
       'Domestic Hotel booking Cashback in Rupees',       'International Hotel booking Cashback in Rupees',
       'Train Booking Cashback in Rupees', 'Minimum Spend', 'Maximum Cashback',
       'Maximum Reward points', 'Reward point VS Cash (x:1)'],
                          index='BankName-CardVariant',aggfunc=np.median)
df_pivot.head()
df_pivot.shape
```

```python
### Scale your variables

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scld = MinMaxScaler(feature_range=(0,1))
arr_scld = scld.fit_transform(df_pivot)

df_scld = pd.DataFrame(arr_scld, columns=df_pivot.columns)

df_scld.head()
df_scld.describe()
df_scld.shape
import warnings
warnings.filterwarnings('ignore')
df_pivot.describe()
```

```python
## Lets build clusters

from sklearn.cluster import KMeans
num_of_clusters = range(2,10)
error = []

for num_clusters in num_of_clusters:
    clusters = KMeans(num_clusters)
    clusters.fit(df_scld)
    error.append(clusters.inertia_/100)
    
df = pd.DataFrame({'Cluster_Numbers':num_of_clusters, "Error_Term":error})
df
```

```python
## Findout the right number of clusters

import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.plot(df.Cluster_Numbers,df.Error_Term, marker = 'D', color='#a51347')

plt.xlabel('Number of Clusters')
plt.ylabel('SSE')     ## Sum of squared error
plt.title('Find the optimal number of clusters')

plt.show()
```

```python
## Build using 6 clusters

clusters = KMeans(8)
clusters.fit(df_scld)

clusters.labels_
```

```python
## centroids
#these are average values in each cluster
clusters.cluster_centers_
```

```python
## Error which is SUM Of Squared Errors

clusters.inertia_
```

```python
df_pivot['Cluster'] = clusters.labels_
clusters.labels_
df_pivot.head()
df_pivot.tail()
df_pivot.shape
df_pivot.to_csv(r'C:\Users\Surendra\OneDrive\Desktop\Capstone\final_cluster_8.csv')
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.style as style

range_n_clusters = [3,4,5,6,7,8,9]
avg_distance=[]
for n_clusters in range_n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=10).fit(df_scld)
    avg_distance.append(clusterer.inertia_)

style.use("fivethirtyeight")
plt.plot(range_n_clusters, avg_distance)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Distance")
plt.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(15,8))
for i in [2, 3, 4, 5]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=50, random_state=10)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(df_scld)
```
